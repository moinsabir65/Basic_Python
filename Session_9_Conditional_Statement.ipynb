```python
##### ask the user enter how much distance you want travel
# assign a variable as user distance
# govt_free distance=25
# if user distance greater than gove free distance
# than only ask user enter charge per 2 km rs
# then calculate total charge
# otherwise print enjoy free ride

import time

user_distance = eval(input('Enter the distance you want to travel: '))
govt_free_distance = 25

if user_distance > govt_free_distance:
    print('Conductor: you have more distance travel to pay charge')
    time.sleep(3)

    print('User: yes sir')

    distance = user_distance - govt_free_distance
    print(f'User: ok sir I will pay for {distance} km')

    charge = eval(input('Enter the charge per km: '))

    total_charge = distance * charge

    print(f'The total payable charge is: {total_charge}')

else:
    print('Enjoy free ride')
```

```python
# wap ask the user enter the percentage
# if percentage greater than 75 then print A grade
# if percentage between 60 to 70 print B grade
# if perncentage between 45 to 60 print c grade
# if percentage between 35 to 45 print D grade
# otherwise print fail

perc=eval(input("Enter the percentage"))  # 90
if(perc >=75 ):       #  90>65  T
    print("Grade A")
elif(perc >=60 ):     #  90>60  T
    print("Grade B")
elif(perc >=45 ):     #  90>45  T
    print("Grade C")
elif perc >=30:       #  90>30  T
    print("Grade D")
else:
    print("You are fail")
```
