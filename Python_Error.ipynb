```python
2. Runtime Error (Exception)

Code sahi likha hai, but run karte waqt problem aayi.

(i)   ZeroDivisionError
	a = 5/0                # 0 se divide? Python bolega: bhai pagal ho kya?

(ii)  NameError
	print(x)               # x define hi nahi kiya. Python bolega: "ye kaun hai bhai?"

(iii) TypeError
	a = "10" + 5           # String + Integer same type hona chahiye

(iv)  ValueError 
	int("abc")             # "abc" ko number banana? Nahi hoga boss.

(v)   IndexError    
	list = [1,2,3]         # List me 5th index hi nahi hai.
        print(list[5])

(vi)  KeyError
	d = {"name":"Moin"}    # Dictionary me "age" key hi nahi hai.
	print(d["age"])

(vii)   AttributeError   
	a = 10                 # Integer ke paas append method nahi hota.
	a.append(5)
```

```python
3. Logical Error	( Sabse dangerous  Error nahi aata, but output galat aata hai. )
	a = 10
	b = 20
	print("Sum =", a-b)   # galti se minus kar diya
```
