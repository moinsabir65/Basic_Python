**Function with arguments**
- in below function we have total 5 varaibles
    - n1,n2,n3,avg and avg1
- input variables are n1,n2 and n3.
- the values provided by user to feed the logic is called as **input variables**
- output variables are : avg,avg1
- output variables are form by input variables
- do not touch **output variables**

```python
def AVERAGE():
    n1=eval(input('Enter a n1:'))
    n2=eval(input('Enter a n2:'))
    n3=eval(input('Enter a n3:'))
    avg=(n1/n2/n3)/3
    avg1=round(avg,2)
    print(f'the avg is {n1} , {n2} and {n3} is : {avg}')

AVERAGE()
```
