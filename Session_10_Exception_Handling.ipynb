**Exception Handling**

- Exception handling is a mechanisum to handle the errors very efficient way

- Usually we will write many number of lines in python code

- Python is a sequential process, if we face an error at one particular line then

- remaining lines will not be executed

- Sometimes the error lines might not be important , it is more often infra related errors

- It does not harmful for our python script

- In this situation we want to continue the execution

- Exception handling catch the errors efficient way with out crash or block the

- **Try Block**
    
    - original code we will write here
    
- **Except Block** 
    - error will catch here
    - if error happend pythone execute the except block code lines

**Try Except Finally**

- try=== original code

- except=== error catching

- finally=== default block
    - finally block always run irrespective of try and except

```python
try:
    print('good')
    print('connection established')
except:
    print('not good')
    print('connection is not established')
finally:
    print('anyways')
    print('connection is lost')
```
